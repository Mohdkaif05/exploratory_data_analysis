## Bivariate analysis

```python
x=df["Area"]
y=df["Fertilizer"]
plt.scatter(x,y)
plt.title("relation b/w fertizer consumption and area")
plt.xlabel("area")
plt.ylabel('fertilizer')
plt.show()
```

```python
x=df["Area"]
y=df["Pesticide"]
plt.scatter(x,y)
plt.title("relation b/w pesticide and area")
plt.xlabel("area")
plt.ylabel("pesticides")
plt.show()
```

#hence both pesticide and fertilizer depend area linearly

```python
x=df["Production"]
y=df["Yield"]
plt.scatter(x,y)
plt.title("relation b/w production and Yield")
plt.xlabel("production")
plt.ylabel("yield")
plt.show()
```

yield show linear relationship with production

```python
x=df["Pesticide"]
y=df["Fertilizer"]
plt.scatter(x,y)
plt.title("relation b/w pesticides and fertilizer")
plt.xlabel("Pesticides")
plt.ylabel("fertilizer")
plt.show()
```

# numerical-categorical data

```python
x=df["Season"]
y=df["Production"]
plt.xlabel("season")
plt.ylabel("production")
plt.title("season wise production")



plt.bar(x,y , width=0.2)
```

```python
x=df["State"]
y=df["Annual_Rainfall"]
plt.bar(x,y)
plt.xlabel("states")
plt.ylabel("annual-rainfall")
plt.title("statewise annual rainfall")
plt.xticks(rotation=90)
plt.show()
```

observation:

```python
x=df["Crop"]
y=df["Fertilizer"]
plt.xticks(rotation=90)
plt.xlabel("crops")
plt.ylabel("amt. of fertilizer")
plt.title("crop wise usage of fertilizer")
plt.bar(x,y)
plt.show()
```

observation:crops like niger seed , wheat , soyabean , rice cotton , gram etc need more fertilizer and pesticides for healthy growth while crop like pulses , spices (cardomom , garlic etc) need less fertilizer and pesticide to grow.

```python
x=df["State"]
y=df["Area"]
plt.xlabel("states")
plt.ylabel("area")
plt.title("statewise area for production of crops")
plt.xticks(rotation=90)
plt.bar(x,y)
plt.show()
```

observation:  In states like West Bengal ,uttar pradesh , madhya pradesh , odisha etc large area is used for cultivation while in states like sikkim , delhi , goa cultivation practice on small area.

although , states in which cultivation practice on large area are also large in size while in states where cultivation practice on small portion of land are small in size.

```python

x=df["Crop"]
y=df["Area"] 
plt.xlabel("crops")
plt.ylabel("area")
plt.title("relation bw crop and area of production")
plt.bar(x,y)
plt.xticks(rotation=90)
plt.show()
```

out of total land area available for cultivation , majority of land used to grow crops like Niger seed, Wheat, Rice,pulses , soyabean etc only.
